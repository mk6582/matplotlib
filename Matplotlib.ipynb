```python
car_sales = pd.read_csv("car_sales.csv")
car_sales
```

```python
ts = pd.Series(np.random.randn(1000),index = pd.date_range("01/01/2022",periods = 1000))

ts = ts.cumsum()
ts.plot();
```

```python
car_sales["Price"]=car_sales["Price"].str.replace('[\$\,\.]','')
car_sales
```

```python
car_sales["Price"]=car_sales["Price"].str[:-2]     ## to remove last two zeros in price colmn
car_sales
```

```python
car_sales["Date"]=pd.date_range("1/1/2022",periods = len(car_sales))
car_sales
```

```python
car_sales["Date"]=car_sales["Date"].dt.strftime('%d-%m-%Y')
car_sales
```

```python
car_sales["Total_sales"] = car_sales["Price"]

car_sales
```

```python
car_sales["Total_sales"] = car_sales["Total_sales"].astype(int).cumsum()

car_sales
```

```python
car_sales.plot(x="Date",y="Total_sales");
```

```python
car_sales["Total_sales"] = car_sales["Price"].astype(int)

car_sales.plot(x="Odometer (KM)",y="Price",kind = "scatter");
```

```python
car_sales.plot(x="Make",y="Odometer (KM)",kind = "bar");
```

```python
heart_disease = pd.read_csv("heart_disease.csv")
heart_disease.head()
```

```python
heart_disease["age"].plot.hist();
```

```python
heart_disease.plot.hist(figsize=(10,30),subplots=True); ## this command use to plot all colums
```

```python
over_50 = heart_disease[heart_disease["age"] > 50]

over_50.head()
```

```python
fig,ax = plt.subplots(figsize=(10,6))
over_50.plot(kind = "scatter",x="age",y= "chol",c="target",ax=ax);
```

```python
fig,ax = plt.subplots(figsize=(10,6))
over_50.plot(kind = "scatter",x="age",y= "chol",c="target",ax=ax);
ax.set_xlim([45,100]);


ax.set(title="Heart Disease",xlabel="Age",ylabel="cholestrol");
```

```python
fig,ax =plt.subplots(figsize=(10,6))
scatter=ax.scatter(x=over_50["age"],
                   y=over_50["chol"],
                 c=over_50["target"] )

ax.set(title="Heart Disease",xlabel="AGE",ylabel="cholestrol");

ax.legend(*scatter.legend_elements(),title="Target");

ax.axhline(over_50["chol"].mean(),linestyle = "--");
```

```python
fig,(ax0,ax1)=plt.subplots(nrows=2,
                          ncols=1,
                          figsize=(10,10))

scatter=ax0.scatter(x=over_50["age"],
                   y=over_50["chol"],
                 c=over_50["target"],
                  cmap="winter");

ax0.set(title="Heart Disease",xlabel="AGE",ylabel="cholestrol");



ax0.legend(*scatter.legend_elements(),title="Target");
ax0.axhline(over_50["chol"].mean(),linestyle = "--");




scatter=ax1.scatter(x=over_50["age"],
                   y=over_50["chol"],
                 c=over_50["target"],
                  cmap="winter");

ax1.set(title="Heart Disease",xlabel="AGE",ylabel="cholestrol");



ax1.legend(*scatter.legend_elements(),title="Target");
ax1.axhline(over_50["chol"].mean(),linestyle = "--")


fig.suptitle('Heart Disease Analysis',fontsize=14,fontweight='bold');
```
